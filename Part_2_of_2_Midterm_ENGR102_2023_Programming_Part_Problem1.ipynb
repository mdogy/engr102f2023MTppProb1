### 1.1 A pseudo-random number sequence

Computers aren't really able to generate truly random numbers. Instead, they use an algoirthm to generate a sequence of numbers that appear to be random. One such algorithm is called the "linear congruential generator". It depends on four parameters: a multiplier we call "a", a modulus we call "m", and an increment we call "c" and a starting value "x0" also called the seed or start. x0 is the first number in the sequence. The next number in the sequence is given by the formula:

$$x_{n+1} = (ax_n + c) \mod{m}$$

where $\mod{}$ is the modulus operator. This operator returns the remainder after dividing the first number by the second. For example, 5 mod 3 is 2 because 5 divided by 3 is 1 with a remainder of 2.
